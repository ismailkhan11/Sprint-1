# *Understanding Variable*

**Loan_ID:** A unique identifier for each loan application.

**Gender:** The gender of the applicant. Possible values are "Male" and "Female."

**Married:** Indicates if the applicant is married or not. Possible values are "Yes" or "No."

**Dependents:** The number of dependents (family members) the applicant has. Possible values are "0", "1", "2", "3+."

**Education:** The education level of the applicant. Possible values are "Graduate" or "Not Graduate."

**Self_Employed:** Indicates whether the applicant is self-employed. Possible values are "Yes" or "No."

**ApplicantIncome:** The income of the applicant.

**CoapplicantIncome:** The income of the co-applicant (if applicable).

**LoanAmount:** The loan amount requested by the applicant.

**Loan_Amount_Term:** The term (duration) of the loan in months.

**Credit_History:** The credit history of the applicant. Possible values are "1" (good credit history) or "0" (no/poor credit history).

**Property_Area:** The area where the property is located. Possible values are "Urban," "Rural," and "Semiurban."

**Loan_Status:** The target variable, indicating whether the loan was approved or not. Possible values are "Y" (Yes) or "N" (No).

# *Data Cleaning*

```python
# It is used to identify and count missing values (NaNs) in a DataFrame
train_data.isnull().sum()
```

**Insight**   
Here we used isnull().sum() to identify and count missing values (NaNs) in a DataFrame

```python
# Fill missing values in categorical columns with mode
train_data['Gender'].fillna(train_data['Gender'].mode()[0], inplace=True)
train_data['Married'].fillna(train_data['Married'].mode()[0], inplace=True)
train_data['Dependents'].fillna(train_data['Dependents'].mode()[0], inplace=True)
train_data['Self_Employed'].fillna(train_data['Self_Employed'].mode()[0], inplace=True)
train_data['Credit_History'].fillna(train_data['Credit_History'].mode()[0], inplace=True)

# Fill missing values in numerical columns with median
train_data['LoanAmount'].fillna(train_data['LoanAmount'].median(), inplace=True)
train_data['Loan_Amount_Term'].fillna(train_data['Loan_Amount_Term'].mode()[0], inplace=True)
```

**Insight**   
We handled missing values in the dataset by replacing them with the mode for categorical columns (e.g., Gender, Married, Dependents, Self_Employed, Credit_History, Loan_Amount_Term) and the median for numerical columns (e.g., LoanAmount). This approach ensures data consistency without introducing unrealistic values like zeros, which could distort the analysis and affect model accuracy.

```python
# Verify that no missing values remain
print(train_data.isnull().sum())
```

**Insight**   
After handling missing values using the mode for categorical columns and the median for numerical columns, we verify that no missing values remain by using isnull().sum(). This ensures the dataset is clean and ready for further analysis and model building.

# **Data Visualization**

# **Visualization 1**

```python
plt.figure(figsize=(8, 6))
palette = ['#3498db', '#e74c3c']
ax = sns.countplot(x='Gender', data=train_data, palette=palette)
total = len(train_data['Gender'])
for p in ax.patches:
    ax.annotate(f'{100 * p.get_height() / total:.1f}%', (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=12)
plt.title('Gender Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

**Insight**     
This plot shows the distribution of male and female applicants. It provides an overview of the gender representation in the dataset.

# **Visualization 2**

```python
plt.figure(figsize=(6, 6))
colors = ['#2ecc71', '#f39c12']
credit_history_counts = train_data['Credit_History'].value_counts()
plt.pie(credit_history_counts, labels=['Good Credit (1)', 'Bad Credit (0)'], autopct='%1.1f%%', startangle=90, colors=colors, textprops={'fontsize': 12})
plt.title('Credit History Distribution', fontsize=16, fontweight='bold')
plt.axis('equal')
plt.show()
```

**Insight**     
This pie chart reveals that a larger percentage of applicants have a good credit history (1), while a smaller proportion have a poor credit history (0). Since credit history is a major factor in loan approval, this distribution indicates that most applicants have a good financial standing.

# **Visualization 3**

```python
# Map 'Y' and 'N' to 'Eligible' and 'Not Eligible'
train_data['Loan_Status'] = train_data['Loan_Status'].map({'Y': 'Eligible', 'N': 'Not Eligible'})

# Create the bar plot with enhanced aesthetics
plt.figure(figsize=(8, 6))
sns.countplot(x='Property_Area', hue='Loan_Status', data=train_data, palette='Set2')
plt.title('Loan Status by Property Area', fontsize=16, fontweight='bold')
plt.xlabel('Property Area', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Loan Status', title_fontsize='13', fontsize='11')
plt.show()
```

**Insight**   
The bar plot reveals how the approval rate of loans varies by property area. This can help determine whether certain property areas are more likely to be approved for loans, which could indicate geographic disparities in loan eligibility. Also, Applicants from semiurban areas have the highest loan approval rates compared to those from urban and rural areas.

# **Visualization 4**

```python
# Mapping 'Y' and 'N' in Loan_Status to 'Eligible' and 'Not Eligible'
train_data['Loan_Status'] = train_data['Loan_Status'].map({'Y': 'Eligible', 'N': 'Not Eligible'})

# Mapping 'Married' to 'Married' and 'Unmarried'
train_data['Married'] = train_data['Married'].map({'Yes': 'Married', 'No': 'Unmarried'})

# Create the count plot for Marital Status and Loan Status
plt.figure(figsize=(8, 6))
sns.countplot(x='Married', hue='Loan_Status', data=train_data, palette ='Set3')  # Updated vibrant colors
plt.title('Loan Status by Marital Status', fontsize=16, fontweight='bold', color='#2F4F4F')
plt.xlabel('Marital Status', fontsize=12, color='#2F4F4F')
plt.ylabel('Count', fontsize=12, color='#2F4F4F')

# Use plt.legend to display the correct labels
plt.legend(title='Loan Status', title_fontsize='13', fontsize='11')

# Display the plot
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Insight**     
This visualization shows that married applicants are more likely to have their loans approved than unmarried applicants. It suggests that marital status might be considered an indicator of stability, influencing loan eligibility decisions.

# Recommendations

**Emphasize Credit History:**  
Since credit history has a significant impact on loan eligibility, banks should prioritize refining how they assess an applicant's creditworthiness. Providing clear guidance to customers on how to maintain or improve their credit history could increase their chances of approval. Educational campaigns on financial responsibility and credit score management can further enhance applicants' profiles.

**Target Specific Loan Products Based on Education:**  
The analysis shows that graduates have higher approval rates. Banks could offer tailored loan products for graduates with more favorable terms. In parallel, they could develop financial education initiatives for non-graduates to improve their loan eligibility.

**Consider Dependents in Loan Offerings:**  
The data reveals that applicants with fewer dependents tend to be approved more often. Banks can develop special loan packages designed for applicants with more dependents, offering more flexible repayment options or lower interest rates to support families with higher financial responsibilities.

**Focus on Semiurban and Rural Areas:**  
Loan approvals are higher in semiurban areas. Banks could use this insight to expand their offerings to urban and rural regions by creating more attractive loan products, such as competitive interest rates or relaxed eligibility criteria, to increase approval rates in those areas.

**Income-Based Loan Packages:** 
Banks can consider designing tiered loan products based on income brackets, offering more affordable loan packages to lower-income applicants while maintaining flexibility for higher-income applicants.

**Improve Financial Literacy for Non-Graduates:**  
Since non-graduates face lower approval rates, banks can help these applicants by offering financial literacy programs, educating them about managing their finances and improving their loan eligibility prospects.

**Data-Driven Decision-Making:**  
Banks can utilize insights from this analysis to make their loan approval process more transparent and data-driven. By focusing on the patterns identified in the data, such as education level, credit history, and dependents, they can streamline their processes and make more informed decisions.
